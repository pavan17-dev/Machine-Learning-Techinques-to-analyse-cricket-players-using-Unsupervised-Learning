```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing dataset
bowling = pd.read_csv("ipl_2022_teams_bat_and_ball.csv")
```

```python
bowling.head()
```

```python
colors = []
# Iterate over rows of ipl_bowlers
for idx, row in bowling.iterrows():
    if row['Team'] == "CSK" :
        colors.append('yellow')
    elif row['Team'] == "RCB" :
        colors.append('green')
    elif row['Team'] == "RR" :
        colors.append('deeppink')
    elif row["Team"] == "DC":
        colors.append('royalblue')
    elif row["Team"] == "GT":
        colors.append("mediumpurple")
    elif row["Team"] == 'PBKS':
        colors.append('red')
    elif row["Team"] == "KKR":
        colors.append('purple')
    elif row["Team"] == "LSG":
        colors.append('skyblue')
    elif row["Team"] == "SRH":
        colors.append("orange")
    elif row["Team"] == "MI":
        colors.append('blue')
    else:
        colors.append('black')
print(colors)
```

```python
# finding the highest strike rate team in batting
# creating bar plot
fig = plt.figure(figsize=(17,9))
x = bowling["Team"]
y = bowling["SR"]
plt.bar(x,y,color=colors,width = 0.8)
plt.title("Batting Strike rate of the teams ipl 2022")
plt.xlabel("Name of the Team")
plt.ylabel("Batting Strike rate of Teams")
plt.show()
```

```python
# finding the highest strike rate team in batting
# creating bar plot
fig = plt.figure(figsize=(17,9))
x = bowling["Team"]
y = bowling["Bowling SR"]
plt.bar(x,y,color=colors,width = 0.8)
plt.title("Bowling Strike rate of the teams ipl 2022")
plt.xlabel("Name of the Team")
plt.ylabel("Bowling Strike rate of Teams")
plt.show()
```

```python
colors = []
# Iterate over rows of ipl_bowlers
for idx, row in bowling.iterrows():
    if row['Team'] == "CSK" :
        colors.append('yellow')
    elif row['Team'] == "RCB" :
        colors.append('green')
    elif row['Team'] == "RR" :
        colors.append('deeppink')
    elif row["Team"] == "DC":
        colors.append('royalblue')
    elif row["Team"] == "GT":
        colors.append("mediumpurple")
    elif row["Team"] == 'PBKS':
        colors.append('red')
    elif row["Team"] == "KKR":
        colors.append('purple')
    elif row["Team"] == "LSG":
        colors.append('skyblue')
    elif row["Team"] == "SRH":
        colors.append("orange")
    elif row["Team"] == "MI":
        colors.append('blue')
    else:
        colors.append('black')
print(colors)
```

```python
# create scatter plot for left arm pacers on there economy and strikerate
fig = plt.figure(figsize=(12,4))
x = bowling["Bowling SR"]
y = bowling["SR"]
plt.scatter(x,y,c=colors,s=100)
plt.title("Bowling Strike Rate and Bowling Strike Rate of of teams in ipl 2022")
plt.xlabel("Bowling Strike Rate of the team")
plt.ylabel("Batting Strike Rate of the team")
for i, label in enumerate(bowling['Team']):
    plt.annotate(label,(x[i], y[i]))
plt.show()
```
