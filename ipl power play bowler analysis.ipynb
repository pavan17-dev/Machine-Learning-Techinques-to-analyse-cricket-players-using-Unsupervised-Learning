## Parts of a DataFrame

To better understand DataFrame objects, it's useful to know that they consist of three components, stored as attributes:

.values: A two-dimensional NumPy array of values.
.columns: An index of columns: the column names.
.index: An index for the rows: either row numbers or row names.

```python
# Print the values of homelessness
print(ipl_power_play_bowlers.values)

# Print the column index of homelessness
print(ipl_power_play_bowlers.columns)

# Print the row index of homelessness
print(ipl_power_play_bowlers.index)
```

## Sorting rows

Finding interesting bits of data in a DataFrame is often easier if you change the order of the rows. You can sort the rows by passing a column name to .sort_values().

In cases where rows have the same value (this is common if you sort on a categorical variable), you may wish to break the ties by sorting on another column. You can sort on multiple columns in this way by passing a list of column names. Sort on … Syntax

```python
# best wicket takers in powerplay
order_of_best_wicket_takers = ipl_power_play_bowlers.sort_values("W", ascending = False)
order_of_best_wicket_takers.head()
```

```python
# best economical bowlers in powerplay
order_of_best_economical_bowlers = ipl_power_play_bowlers.sort_values("Econ")
order_of_best_economical_bowlers.head(5)
```

```python
# best wicket takers and economical bowlers in powerplay overs
best_bowlers_in_powerplay = ipl_power_play_bowlers.sort_values(["W","Econ"],ascending = [False,True])
best_bowlers_in_powerplay.head(5)
```

## Subsetting columns

When working with data, you may not need all of the variables in your dataset. Square brackets ([]) can be used to select only the columns that matter to you in an order that makes sense to you. To select only "col_a" of the DataFrame df, use

df["col_a"]

To select "col_a" and "col_b" of df, use

df[["col_a", "col_b"]]


```python
ipl_power_play_bowlers_df = ipl_power_play_bowlers[["Player","Team",'I',"W",'Avg','SR','Econ','BP4G','BPB','Fours %',
                                                    'bounadry %','Dot Ball %','NBECO']]
```

## Subsetting rows

A large part of data science is about finding which bits of your dataset are interesting. One of the simplest techniques for this is to find a subset of rows that match some criteria. This is sometimes known as filtering rows or selecting rows.

There are many ways to subset a DataFrame, perhaps the most common is to use relational operators to return True or False for each row, then pass that inside square brackets.

dogs[dogs["height_cm"] > 60]

dogs[dogs["color"] == "tan"]

```python
rr_power_play_bolwers = ipl_power_play_bolwers_df[ipl_power_play_bolwers_df["Team"] == "RR"]
rr_power_play_bolwers.head(2)
```

```python
# ipl powerplay bowlers who have economy < 7
economical_bowlers_in_pp = ipl_power_play_bolwers_df[ipl_power_play_bolwers_df["Econ"] <= 7.00]
print(economical_bowlers_in_pp)
```

```python
# ipl economical and good strike rate bowlers in power play
good_bowlers_in_pp = ipl_power_play_bolwers_df[(ipl_power_play_bolwers_df["Econ"] < 8.00) & 
                                               (ipl_power_play_bolwers_df["SR"] < 24)]
```

```python
good_bowlers_in_pp = good_bowlers_in_pp.reset_index()
```

```python
print(good_bowlers_in_pp)
```

```python
colors = []
# Iterate over rows of ipl_bowlers
for idx, row in good_bowlers_in_pp.iterrows():
    if row['Team'] == "CSK" :
        colors.append('yellow')
    elif row['Team'] == "RCB" :
        colors.append('green')
    elif row['Team'] == "RR" :
        colors.append('deeppink')
    elif row["Team"] == "DC":
        colors.append('royalblue')
    elif row["Team"] == "GT":
        colors.append("mediumpurple")
    elif row["Team"] == 'PBKS':
        colors.append('red')
    elif row["Team"] == "KKR":
        colors.append('purple')
    elif row["Team"] == "LSG":
        colors.append('skyblue')
    elif row["Team"] == "SRH":
        colors.append("orange")
    elif row["Team"] == "MI":
        colors.append('blue')
    else:
        colors.append('black')
print(colors)
```

```python
# create scatter plot for right arm pacers on there economy and strikerate
fig = plt.figure(figsize=(12,4))
x = good_bowlers_in_pp["Econ"]
y = good_bowlers_in_pp["SR"]
plt.scatter(x,y,c=colors,s=100)
plt.title("Economy and strike rate of Power Play Bowlers in ipl 2022")
plt.xlabel("Economy of the player")
plt.ylabel("Bowling strike rate of the player")
for i, label in enumerate(good_bowlers_in_pp['Player']):
    plt.annotate(label,(x[i], y[i]))
plt.show()
```

## Subsetting rows by categorical variables

Subsetting data based on a categorical variable often involves using the "or" operator (|) to select rows from multiple categories. This can get tedious when you want all states in one of three different regions, for example. Instead, use the .isin() method, which will allow you to tackle this problem by writing one condition instead of three separate ones.

colors = ["brown", "black", "tan"]

condition = dogs["color"].isin(colors)

dogs[condition]

## Adding new columns

You aren't stuck with just the data you are given. Instead, you can add new columns to a DataFrame. This has many names, such as transforming, mutating, and feature engineering.

You can create new columns from scratch, but it is also common to derive them from other columns, for example, by adding columns together or by changing their units.

```python
ipl_power_play_bowlers_df["WPI"] = (ipl_power_play_bowlers_df['W'] / ipl_power_play_bowlers_df["I"])
print(ipl_power_play_bowlers_df)
```

```python
ipl_power_play_bolwers_df["Agg_Avg_Str"] = ipl_power_play_bolwers_df['Avg']*0.35 + ipl_power_play_bolwers_df["SR"]*0.75
print(ipl_power_play_bolwers_df)
```

## Combo-attack!

You've seen the four most common types of data manipulation: sorting rows, subsetting columns, subsetting rows, and adding new columns. In a real-life data analysis, you can mix and match these four manipulations to answer a multitude of questions.

```python
# adding WPI and agg_avg_str columns to ipl_power_play_bowlers

ipl_power_play_bowlers["WPI"] = (ipl_power_play_bowlers['W'] / ipl_power_play_bowlers["I"])
#print(ipl_power_play_bowlers)

ipl_power_play_bowlers["Agg_Avg_Str"] = ipl_power_play_bowlers['Avg']*0.35 + ipl_power_play_bowlers["SR"]*0.75

# subsetting important columns for analysis
ipl_power_play_bowlers_df = ipl_power_play_bowlers[["Player","Team",'I',"W",'Avg','SR','Econ','BP4G','BPB','Fours %',
                                                    'bounadry %','Dot Ball %','NBECO',"WPI","Agg_Avg_Str"]]
#print(ipl_power_play_bowlers_df)

# sorting the players with WPI, Agg_Avg_Str 
ipl_power_play_bowlers_df_srt = ipl_power_play_bowlers_df.sort_values(["WPI","Agg_Avg_Str"],ascending = [False,True])
#print(best_bowlers_in_pp_srt)

# subsetting rows who have wpi > 0.53 and Agg_Avg_Str < 30
best_bowlers_in_power_play = ipl_power_play_bowlers_df[(ipl_power_play_bowlers_df["WPI"] > 0.50) & (ipl_power_play_bowlers_df
                                                                                            ["Agg_Avg_Str"]) ]
#print(best_bowlers_in_power_play)


# sorting best WPI Bowlers
best_wicket_taking_bowler_in_power_play = best_bowlers_in_power_play.sort_values(["WPI","Agg_Avg_Str"],ascending = [False,True])
#print(best_wicket_taking_bowler_in_power_play.head())

# sorting best 5 powerplay bowlers in ipl
atlast_best_bowlers_in_pp = best_bowlers_in_power_play.sort_values(["Agg_Avg_Str","W"],ascending = [True,False])

print(atlast_best_bowlers_in_pp.head())
```
